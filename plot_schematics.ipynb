```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# Parameters
circle_radius = 1
theta = np.pi/4  # 45 degrees
phi = np.pi/6    # 30 degrees

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(6,6))

# Plot Circle
circle = plt.Circle((0, 0), circle_radius, color='b', fill=False)
ax.add_artist(circle)

# Plot Red Dot
x_dot = circle_radius * np.cos(theta)
y_dot = circle_radius * np.sin(theta)
plt.plot(x_dot, y_dot, 'ro')

# Draw a line from the center to the red dot
plt.plot([0, x_dot], [0, y_dot], 'r-')

# Draw Theta Arc
theta_arc = patches.Arc((0,0), 0.4, 0.4, 
                        theta1=0, theta2=45, 
                        color='red', linestyle='dashed')
ax.add_patch(theta_arc)
plt.annotate(r'$\theta$', (0.2, 0.1), fontsize=12, color='red')

# Draw Arrows and Annotate Phi
arrow_positions = [(-1.5, -1.5), (-1.7, -1.3), (-1.9, -1.1)]  # Bottom left side
for pos in arrow_positions:
    dx = np.cos(phi)  # dx and dy are components of the arrow direction
    dy = np.sin(phi)
    plt.arrow(pos[0], pos[1], dx, dy, 
              shape='full', lw=0, length_includes_head=True, 
              head_width=0.1, color='green')
    
    # Draw Phi Arc
    phi_arc = patches.Arc(pos, 0.4, 0.4, 
                          theta1=0, theta2=30, 
                          color='green', linestyle='dashed')
    ax.add_patch(phi_arc)
    
    # Annotate Phi (only once near the first arrow for clarity)
    if pos == arrow_positions[0]:
        plt.annotate(r'$\phi$', (pos[0]+0.2, pos[1]), 
                     fontsize=12, color='green')

# Set equal scaling and show the plot
ax.axis('equal')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.title('Circle with Angles and Arrows')
plt.savefig('schematic.jpg',dpi=600)
plt.show()
```
